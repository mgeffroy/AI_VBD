```python
import pandas as pd
import requests
import itertools
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Base URL for PubMed API
base_url = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi'
```

```python
# Lists of keywords
diseases = ['Dengue', 'Zika', 'Chikungunya','lymphatic filariasis', 'rift valley fever', 'yellow fever', 'malaria', 'japanese encephalitis',
            'west nile fever', 'Schistosomiasis','onchocerciasis','plague','tungiasis','typhus','leishmania','sandfly fever','lyme',
            'borrelia','rickettsia','tularemia','chagas','sleeping sickness']
ai_types = ['deep learning','machine learning', 'k-means clustering','logistic regression','support vector machine',
            'text mining', 'artificial neural networks', 'expert system','random forest', 'decision trees', 
            'gradient boosting', 'natural language processing', 'sentiment analysis','fuzzy logic']
objectives = ['surveillance', 'forecast','early warning system', 'diagnostic']
```

```python
# Pathogen type attribute dictionary
disease_pathogen = {
    'Dengue': 'virus',
    'Zika': 'virus',
    'Chikungunya': 'virus',
    'lymphatic filariasis':'parasite', 
    'rift valley fever': 'virus', 
    'yellow fever': 'virus', 
    'malaria':'parasite',
    'japanese encephalitis':'virus',
    'west nile fever': 'virus',
    'Schistosomiasis':'parasite',
    'onchocerciasis':'parasite',
    'plague':'bacteria',
    'tungiasis':'ectoparasite',
    'typhus':'bacteria',
    'leishmania':'parasite',
    'sandfly fever':'virus',
    'lyme':'bacteria',
    'borrelia':'bacteria',
    'rickettsia':'bacteria',
    'tularemia':'bacteria',
    'chagas':'parasite',
    'sleeping sickness':'parasite'
}
```

```python
# Neglected disease attribute dictionary
neglected_status = {
    'Dengue': 'Y',
    'Zika': 'N',
    'Chikungunya': 'Y',
    'lymphatic filariasis':'Y', 
    'rift valley fever': 'N', 
    'yellow fever': 'Y', 
    'malaria':'Y',
    'japanese encephalitis':'N',
    'west nile fever': 'N',
    'Schistosomiasis':'Y',
    'onchocerciasis':'Y',
    'plague':'N',
    'tungiasis':'N',
    'typhus':'N',
    'leishmania':'Y',
    'sandfly fever':'N',
    'lyme':'N',
    'borrelia':'N',
    'rickettsia':'N',
    'tularemia':'N',
    'chagas':'Y',
    'sleeping sickness':'Y'
}
```

```python
#TYpe of vector attribute
vector_type= {
    "Dengue": 'mosquito', 
    'Zika': 'mosquito',
    'Chikungunya': 'mosquito',
    'lymphatic filariasis':'mosquito', 
    'rift valley fever': 'mosquito', 
    'yellow fever': 'mosquito', 
    'malaria':'mosquito',
    'japanese encephalitis':'mosquito',
    'west nile fever': 'mosquito',
    'Schistosomiasis':'aquatic snail',
    'onchocerciasis':'blackfly',
    'plague':'fleas',
    'tungiasis':'fleas',
    'typhus':'Lice',
    'leishmania':'Sandfly',
    'sandfly fever':'Sandfly',
    'lyme':'ticks',
    'borrelia':'ticks',
    'rickettsia':'ticks',
    'tularemia':'ticks',
    'chagas':'triatome bugs',
    'sleeping sickness':'tsetse flies'

    
}
```

```python
#Generate combinations of diseases, AI types and objectives

keyword_combinations = list(itertools.product(diseases, ai_types, objectives))

#generate a emtpy list to store results 
results =[]

for disease, ai_type, objective in keyword_combinations:
    query = f'{disease} AND {ai_type} AND {objective}'
    try:
        # Set up the parameters for the PubMed API request
        params = {
            'db': 'pubmed',
            'term': query,
            'retmode': 'json',
            'retmax': 0, 
            'datetype':'pdat',
            'mindate': '2000',
            'maxdate':'2024'
        }
        # Make the API request
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        data = response.json()
        result_count = int(data['esearchresult']['count'])
        
        # Append the results to the list
        results.append({
            'Disease': disease,
            'Vector': vector_type[disease],
            'Attribute': disease_pathogen[disease],
            'Neglected': neglected_status[disease],
            'AI_Type': ai_type,
            'Objective': objective,
            'Query': query,
            'Result_Count': result_count
        })
    except Exception as e:
        print(f"Error with query '{query}': {e}")
        results.append({
            'Disease': disease,
            'Vector': vector_type[disease],
            'Attribute': disease_pathogen[disease],
            'Neglected': neglected_status[disease],
            'AI_Type': ai_type,
            'Objective': objective,
            'Query': query,
            'Result_Count': 'Error'
        })

# Convert the results list to a DataFrame
df = pd.DataFrame(results)

# Display the DataFrame
print(df)
```

```python
# Pivoting the data for heatmap
heatmap_data = df.pivot_table(index=['AI_Type'], columns=['Disease', 'Objective'], values='Result_Count', aggfunc='sum')

# Plotting the heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Number of Results'})

# Customizing the plot
plt.title("Number of PubMed Results for Different AI Techniques by Disease and Objective")
plt.xlabel("Disease and Objective")
plt.ylabel("AI Technique")
plt.show()
```
